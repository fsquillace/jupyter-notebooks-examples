# Histograms
By default, Python sets the number of bins to 10 in that case. The number of bins is pretty important. Too few bins will oversimplify reality and won't show you the details. Too many bins will overcomplicate reality and won't show the bigger picture.

```python
_, axes_hist = plt.subplots()

# Build histogram with 5 bins
axes_hist.hist(life_exp, bins=5, color='cyan')

# Build histogram with 20 bins
axes_hist.hist(life_exp, bins=20, color='red')
```

# Subplot
`plt.subplot(m, n, k)` will make the subplot grid of dimensions `m` by `n` and to make the `k`th subplot active (subplots are numbered starting from 1 row-wise from the top left corner of the subplot grid).

```python
# Create a figure with 1x2 subplot and make the left subplot active
plt.subplot(2, 2, 1)

plt.hist(life_exp, bins=5, color='blue')
plt.title('Life exp on 5 bins')

# Make the right subplot active in the current 1x2 subplot grid
plt.subplot(2, 2, 2)

plt.hist(life_exp, bins=20, color='red')
plt.title('Life exp on 20 bins')

# Make the right subplot active in the current 1x2 subplot grid
plt.subplot(2, 2, 3)

plt.hist(life_exp, bins=40, color='green')
plt.title('Life exp on 40 bins')

# Make the right subplot active in the current 1x2 subplot grid
plt.subplot(2, 2, 4)

plt.hist(life_exp, bins=60, color='yellow')
plt.title('Life exp on 60 bins')

# Use plt.tight_layout() to improve the spacing between subplots
plt.tight_layout()
plt.show()
```
