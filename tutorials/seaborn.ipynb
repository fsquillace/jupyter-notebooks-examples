## Figure Aesthetics

### Seaborn styles

```python
#(Re)set the seaborn default
sns.set()                         

# Set the matplotlib parameters
sns.set_style("whitegrid")

graph = sns.lmplot(x='high-speed-running', y='distance', data=weekDf)

plt.show(graph)
```

```python
#(Re)set the seaborn default
sns.set()

sns.set_style("ticks", {"xtick.major.size":8,"ytick.major.size":8})

# Return a dict of params or use with `with` to temporarily set the style
sns.axes_style("darkgrid")

graph = sns.lmplot(x='high-speed-running', y='distance', data=weekDf)
plt.show(graph)
```

### Context functions

```python
#(Re)set the seaborn default
sns.set()

sns.set_context("talk")

graph = sns.lmplot(x='high-speed-running', y='distance', data=weekDf)
plt.show(graph)
```

```python
#(Re)set the seaborn default
sns.set()

sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth":2.5})
graph = sns.lmplot(x='high-speed-running', y='distance', data=weekDf)
plt.show(graph)
```

### Color palette

```python
#(Re)set the seaborn default
sns.set()

sns.set_palette("husl",3)
graph = sns.lmplot(x='high-speed-running', y='distance', data=weekDf)
plt.show(graph)
```

```python
#(Re)set the seaborn default
sns.set()
# Set your own color palette
flatui = ["#9b59b6","#3498db","#95a5a6","#e74c3c","#34495e","#2ecc71"]
sns.set_palette(flatui)         

graph = sns.lmplot(x='high-speed-running', y='distance', data=weekDf)
plt.show(graph)
```

### Axisgrid objects

```python
sns.set()

g = sns.lmplot(x='high-speed-running', y='distance', data=weekDf)

# Remove left spine 
g.despine(left=True)

# Set the labels of the y-axis
g.set_ylabels("space")


# Set the tick labels for x
g.set_xticklabels(rotation=45)

# Set the axis labels
g.set_axis_labels("speed", "space")

# Set the limit and ticks of the x-and y-axis
g.set(xlim=(3,4), ylim=(18,35), xticks=[3,0.5,4], yticks=[18,0.5,35])

plt.show(g)
```

## Plotting with Seaborn

### Axis Grids

```python
#(Re)set the seaborn default
sns.set()

# Subplot grid for plotting conditional relationships
g = sns.FacetGrid(titanic, col="survived",row="sex")
g = g.map(plt.hist,"age")

plt.show(g)
```

```python
#(Re)set the seaborn default
sns.set()

# Draw a categorical plot onto a Facetgrid
g = sns.factorplot(x="pclass", y="survived", hue="sex", data=titanic) 

plt.show(g)
```

```python
#(Re)set the seaborn default
sns.set()

# Plot data and regression model fits across a FacetGrid
g = sns.lmplot(x="sepal_width", y="sepal_length", hue="species", data=iris)

plt.show(g)
```

```python
#(Re)set the seaborn default
sns.set()

# Subplot grid for plotting pairwise relationships
h = sns.PairGrid(iris)
h = h.map(plt.scatter)

plt.show(h)
```

```python
#(Re)set the seaborn default
sns.set()

# Plot pairwise bivariate distributions
h = sns.pairplot(iris)

plt.show(h)
```

```python
#(Re)set the seaborn default
sns.set()

# Grid for bivariate plot with marginal univariate plots 
i = sns.JointGrid(x="x", y="y", data=data)
i = i.plot(sns.regplot, sns.distplot)

plt.show(i)
```

```python
#(Re)set the seaborn default
sns.set()

# Plot bivariate distribution
g = sns.jointplot("sepal_length", "sepal_width", data=iris, kind='kde')

plt.show(g)
```

### Regression plots

```python
#(Re)set the seaborn default
sns.set()

# Plot data and a linear regression model fit
g = sns.regplot(x="sepal_width", y="sepal_length", data=iris)    

plt.show(g)
```

### Distribution plots

```python
#(Re)set the seaborn default
sns.set()

# Plot univariate distribution
g = sns.distplot(data.y, kde=False, color="b")

plt.show(g)
```

### Matrix plots

```python
#(Re)set the seaborn default
sns.set()

# Heatmap
g = sns.heatmap(uniform_data,vmin=0,vmax=1)

plt.show(g)
```

### Categorical plots

```python
#(Re)set the seaborn default
sns.set()

# Scatterplot

# Scatterplot with one categorical variable
sns.stripplot(x="species", y="petal_length", data=iris) 

plt.show(g)
```

```python
#(Re)set the seaborn default
sns.set()

# Categorical scatterplot with non-overlapping points
g = sns.swarmplot(x="species", y="petal_length", data=iris)

plt.show(g)
```

```python
sns.set()

# Bar Chart

# Show point estimates and confidence intervals with scatterplot glyphs
g = sns.barplot(x="sex", y="survived", hue="class", data=titanic)


plt.show(g)
```

```python
sns.set()

# Count Plot
# Show count of observations
g = sns.countplot(x="deck", data=titanic, palette="Greens_d")

plt.show(g)
```

```python
sns.set()

# Point Plot 
# Show point estimates and confidence intervals as rectangular bars
g = sns.pointplot(x="class", y="survived", hue="sex", data=titanic,
                  palette={"male":"g", "female":"m"}, markers=["^", "o"], linestyles=["-","--"])

plt.show(g)
```

```python
sns.set()

# Boxplot
g = sns.boxplot(x="alive", y="age", hue="adult_male", data=titanic)

plt.show(g)
```

```python
sns.set()

# Boxplot with wide-form data
g = sns.boxplot(data=iris, orient="h")

plt.show(g)
```

```python
sns.set()


# Violinplot
g = sns.violinplot(x="age", y="sex", hue="survived", data=titanic)

plt.show(g)
```
